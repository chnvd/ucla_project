## Year

Some years are non-numeric, hence the dataype is object. We want this to be uniform.
There are 
- 213 non-numeric end_year
- 152 end_years after 2023
- 86 non-numeric start_year
- 41 of this 86 non-numeric start years have a numeric end year
- There are 168 numeric start_year values among the non-numeric end years

So, to impute some of these values, I find the average number of years projects have been in existance for each of the country with the missing values, and use the numberic start (end) date and arrive at the end (start) data by subtracting (adding) the average duration of construction. 

```python
# Finding # rows with the issues: 
# non-numeric dates
def non_numeric(col):
    # Identify non-numeric values in the column
    nn = transit_cost[col].apply(lambda x: not str(x).isnumeric())
    print(f"There are {nn.sum()} non-numeric values in {col}")
    transit_cost[nn].head()
non_numeric('start_year')
non_numeric('end_year')
```

```python
# Identify numeric values in the 'End year' column which are greater than 2023
nn_23 = transit_cost['end_year'].apply(lambda x: int(x)>2023 if str(x).isnumeric() else False)
# Count the number of such values
print(f"There are {nn_23.sum()} end_year values > 2023")
```

```python
# identify # rows with end year value but no start year value
nns_nostart = transit_cost['start_year'].apply(lambda x: not str(x).isnumeric()) & transit_cost['end_year'].apply(lambda x: str(x).isnumeric())
print(f"There are {nns_nostart.sum()} numeric end_year values among the non-numeric start years")
# list of countries with this pheonomeon
print("Countries:", transit_cost[nns_nostart].country.unique())
# identify # rows with start year value but no end year value
nns_noend = transit_cost['end_year'].apply(lambda x: not str(x).isnumeric()) & transit_cost['start_year'].apply(lambda x: str(x).isnumeric())
print(f"There are {nns_noend.sum()} numeric start_year values among the non-numeric end years")
# list of countries with this pheonomeon
print("Countries:", transit_cost[nns_noend].country.unique())
```

Let us find average built duration for CN, IR, IT and use that to impute start years. 

```python
def avg_dur(country):
    data = transit_cost[transit_cost['country']==country]
    # retaining numeric dates
    numeric = data['start_year'].apply(lambda x: str(x).isnumeric())
    data = data[numeric]
    numeric = data['end_year'].apply(lambda x: str(x).isnumeric())
    data = data[numeric]
    # calculate average duration
    data['start_year'] = pd.to_numeric(data['start_year'], errors='coerce')
    data['end_year'] = pd.to_numeric(data['end_year'], errors='coerce') 
    data['duration'] = data['end_year'] - data['start_year']
    average_duration = data['duration'].mean().round().astype(int)
    # print(f"The average duration of the projects is {average_duration} years for {country}")
    return average_duration
print(f"The average duration of the projects is {avg_dur('CN')} years for CN")
print(f"The average duration of the projects is {avg_dur('IR')} years for IR")
print(f"The average duration of the projects is {avg_dur('IT')} years for IT")
```

```python
# calcuate average duration for all countries and save into a df - not immediately useful.
country_df = pd.DataFrame(transit_cost['country'].unique())
country_df.rename(columns={0: 'country'}, inplace=True)
country_df['duration'] = country_df['country'].apply (lambda x: avg_dur(x))
country_df.head()
```

```python
# imputing start years
nns_nostart = transit_cost['start_year'].apply(lambda x: not str(x).isnumeric()) & transit_cost['end_year'].apply(lambda x: str(x).isnumeric())
# condition = transit_cost['start_year'].apply(lambda x: not str(x).isnumeric())
pd.to_numeric(transit_cost['end_year'], errors='coerce')
if nns_nostart.any():
    transit_cost.loc[nns_nostart,'start_year'] = transit_cost['end_year'] - transit_cost.loc[nns_nostart, 'country'].apply(avg_dur)

#imputing end years
nns_noend = transit_cost['end_year'].apply(lambda x: not str(x).isnumeric()) & transit_cost['start_year'].apply(lambda x: str(x).isnumeric())
pd.to_numeric(transit_cost['start_year'], errors='coerce')
if nns_noend.any():
    transit_cost.loc[nns_noend,'end_year'] = transit_cost['start_year'] + transit_cost.loc[nns_noend, 'country'].apply(avg_dur)
```

## Tunnel details and number of station

1. convert places with no tunnels - assign city wise avegate tunnel perc and multiply by the lenght of line for which tunnel is missing
2. for station# missing = find countrywise avg station per km of line, ans assign to missing row *length of the row


```python
print(transit_df.isnull().sum())
```

```python
def non_numeric(col):
    # Identify non-numeric values in the column
    nn = transit_df[col].isnull()
    print(f"There are {nn.sum()} non-numeric values in {col}")
    return nn
df = non_numeric('tunnelper')

# print(f'There are {transit_df[df].city.nunique()} are in the following cities:', transit_df[df].city.unique())
tunnelgroup = transit_df.groupby('city')['tunnelper'].mean()

# this cleans 
for index, row in transit_df[df].iterrows():
    transit_df.loc[index, 'tunnel'] =(tunnelgroup[transit_df.loc[index, 'city']])* transit_df.loc[index,'length'] 
    transit_df.loc[index, 'tunnelper'] = transit_df.loc[index, 'tunnel']/transit_df.loc[index, 'length']

non_numeric('tunnelper')

transit_df = transit_df[transit_df['tunnelper'].notnull()]

non_numeric('tunnelper')
```

```python
print(transit_df.isnull().sum())
```

```python
def no_station(col):
    # Identify non-numeric values in the column
    nn = transit_df[col].isnull()
    print(f"There are {nn.sum()} non-numeric values in {col}")
    return nn
    # print(transit_df[nn])
stations = no_station('stations')
stationdf = transit_df.groupby('country')[['length','stations']].mean()
# there is one country with missing station data for all its projects - Qatar. We find the number as available online replace the value there
stationdf.at['QA','stations']=37
# finding number of stations per km
stationdf['st_km']= (stationdf['stations']/stationdf['length'])
```

```python
for index, row in transit_df[stations].iterrows():
    transit_df.at[index,'stations'] =  row['length'] * stationdf.loc[row['country'], 'st_km']
print(transit_df.isnull().sum())
```
